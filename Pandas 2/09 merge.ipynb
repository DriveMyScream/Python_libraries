```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'city': ["new york", "chicago", "orlando"],
    'temperature': [21, 14, 35]
})
df1
```

```python
df2 = pd.DataFrame({
    'city': ["orlando", "chicago", "new york"],
    'humidity': [40, 30, 50]
})
df2
```

```python
newdf = pd.merge(df1, df2, on="city")
newdf
```

```python
df1 = pd.DataFrame({
    'city': ["new york", "chicago", "orlando", "california"],
    'temperature': [30, 25, 30, 26]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["washington dc", "new york", "california", "Los Angeles"],
    "Humidity": [30, 45, 50, 36]
})
df2
```

```python
newdf = pd.concat([df1, df2], ignore_index=True)
newdf
```

```python
newdf = pd.merge(df1, df2, on="city", how='inner')
newdf
```

```python
newdf = pd.merge(df1, df2, on="city", how='outer')
newdf
```

```python
newdf = pd.merge(df1, df2, on="city", how="left")
newdf
```

```python
newdf = pd.merge(df1, df2, on="city", how="right")
newdf
```

```python
newdf = pd.merge(df1, df2, on="city", how="outer", indicator=True)
newdf
```

### Common columns in data frame

```python
df1 = pd.DataFrame({
    "city": ["chicago", "new york", "Los Angeles"],
    "humidity": [40, 50, 30],
    "temperature": [30, 35, 40]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["Los Angeles", "washington dc", "chicago"],
    "humidity": [30, 40, 50],
    "temperature": [35, 40, 45]
})
df2
```

```python
newdf = pd.concat([df1, df2])
newdf
```

```python
newdf = pd.merge(df1, df2, on="city")
newdf
```

```python
newdf = pd.merge(df1, df2, on="city", suffixes=('_first', '_second'))
newdf
```

### Join

```python
df1 = pd.DataFrame({
    "city": ["chicago", "Los Angeles", "new york"],
    "temperature": [30, 35, 40]
})
df1.set_index('city', inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["Los Angeles", "washington dc", "new york"],
    "humidity": [30, 35, 40]
})
df2.set_index('city', inplace=True)
df2
```

```python
newdf = df1.join(df2)
newdf
```
