```python
df = pd.read_csv("weather_data.csv", parse_dates=['day'])
```

```python
df
```

```python
print("The type of date in day is:", type(df["day"][0]))
```

```python
df.set_index("day", inplace=True)
```

```python
df
```

```python
newdf = df.fillna(0)
newdf
```

```python
newdf = df.fillna(df.mean())
newdf
```

```python
newdf = df.fillna({"temperature": df["temperature"].mean(),
                   "windspeed": df["windspeed"].mean(),
                   "event": df["event"].mode()})
```

```python
newdf
```

```python
newdf = df.fillna(method='ffill')
newdf
```

```python
newdf = df.fillna(method='bfill')
newdf
```

```python
newdf = df.fillna(df.mean(), limit=2)
newdf
```

```python
newdf = df.interpolate()
newdf
```
