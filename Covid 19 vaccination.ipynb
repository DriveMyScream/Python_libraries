### Adding more column in the data set
* This column's is used for daily, weekly and monthly analysis perpose.
1. year
2. month
3. day
4. weekday

```python
vaccine_df['year'] = pd.DatetimeIndex(vaccine_df['date']).year
vaccine_df['month'] = pd.DatetimeIndex(vaccine_df['date']).month
vaccine_df['day'] = pd.DatetimeIndex(vaccine_df['date']).day
vaccine_df['weekday'] = pd.DatetimeIndex(vaccine_df['date']).weekday
```

```python
vaccine_df
```

### Adding more columns into the data set
* This columns is the total sum of columns
* This columns is used for total result of that column perpose
1. total_new_cases
2. total_new_deaths
3. total_new_tests
4. total_total_vaccinations
5. total_people_vaccinated
6. total_people_fully_vaccinated

```python
vaccine_df['total_new_cases'] = vaccine_df['new_cases'].cumsum()
vaccine_df['total_new_deaths'] = vaccine_df['new_deaths'].cumsum()
vaccine_df['total_new_tests'] = vaccine_df['new_tests'].cumsum()
vaccine_df['total_total_vaccinations'] = vaccine_df['total_vaccinations'].cumsum()
vaccine_df['total_people_vaccinated'] = vaccine_df['people_vaccinated'].cumsum()
vaccine_df['total_people_fully_vaccinated'] = vaccine_df['people_fully_vaccinated'].cumsum()
```

```python
vaccine_df
```

### Adding more columns into the data set
* This columns is the columns with respect to their million.
* This columns is for million wise analysis perpose.
1. new_cases_per_million
2. new_deaths_per_million
3. new_tests_per_million
4. total_vaccinations_per_million
5. people_vaccinated_per_million
6. people_fully_vaccinated_per_million

```python
vaccine_df['new_cases_per_million'] = vaccine_df['new_cases'] * 1e6 / vaccine_df['population']
vaccine_df['new_deaths_per_million'] = vaccine_df['new_deaths'] * 1e6 / vaccine_df['population']
vaccine_df['new_tests_per_million'] = vaccine_df['new_tests'] * 1e6 / vaccine_df['population']
vaccine_df['total_vaccinations_per_million'] = vaccine_df['total_vaccinations'] * 1e6 / vaccine_df['population']
vaccine_df['people_vaccinated_per_million'] = vaccine_df['people_vaccinated'] * 1e6 / vaccine_df['population']
vaccine_df['people_fully_vaccinated_per_million'] = vaccine_df['people_fully_vaccinated'] * 1e6 / vaccine_df['population']
```
