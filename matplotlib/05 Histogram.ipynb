```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
blood_suger = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_suger, rwidth=0.8);
```

```python
plt.hist(blood_suger, rwidth=0.5, bins=4);
```

```python
plt.xlabel("Diabetic patients distrubution")
plt.ylabel("Count of values")
plt.title("Diabetic patients data")
plt.hist(blood_suger, rwidth=0.95, bins=[80, 100, 125, 150]);
```

#### Multiple histogram

```python
blood_suger_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_suger_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Suger level")
plt.ylabel("Number of patients")
plt.title("Blood suger chart")
plt.hist([blood_suger_men, blood_suger_women], rwidth=0.95, bins=[80, 100, 125, 150], label=['Men', 'Women'])
plt.legend();
```

```python
plt.xlabel("Suger level")
plt.ylabel("Number of patients")
plt.title("Blood suger chart")
plt.hist(blood_suger_men, rwidth=0.95, histtype='step');
```

```python
plt.xlabel("Suger lavel")
plt.ylabel("Number of patients")
plt.title("Blood suger chart")
plt.hist([blood_suger_men, blood_suger_women], label=['Men', 'Women'], bins=[80, 100, 125, 150], rwidth=0.95, orientation='horizontal');
plt.legend();
```
