```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thusday", "Friday", "Saturday"]
max_temp = [50, 55, 53, 57, 58, 50, 58]
min_temp = [30, 33, 37, 39, 36, 38, 34]
mean_temp = [40, 43, 47, 49, 48, 43, 41]
```

```python
plt.plot(days, max_temp, label="Maximum temp")
plt.plot(days, mean_temp, label="Mean tempe")
plt.plot(days, min_temp, label="Minimum temp");
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temperature in different days")
plt.plot(days, min_temp, label="Manimum temp")
plt.plot(days, mean_temp, label="Mean temp")
plt.plot(days, max_temp, label="Maximum temp")
plt.legend();
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temperature in different days")
plt.plot(days, min_temp, label="Min temp")
plt.plot(days, mean_temp, label="Mean temp")
plt.plot(days, max_temp, label="Max temp")
plt.legend(shadow=True, fontsize="large");
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temperature in different days")
plt.plot(days, min_temp, label="Min temp")
plt.plot(days, mean_temp, label="Mean temp")
plt.plot(days, max_temp, label="Max temp")
plt.legend(shadow=True, loc="best", fontsize="large")
plt.grid();
```
