```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home rent", "Food", "Phone", "Car", "Other"]
plt.pie(exp_vals, labels=exp_labels);
```

#### For perfect circle

```python
plt.pie(exp_vals, labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show();
```

#### Show percentage for every pie. specify radius to increase chart size

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct="%.2f", radius=1.5)
plt.show();
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct="%.2f", radius=1.5, explode=[0, 0, 0, 0.1, 0.2])
plt.show();
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct="%.2f", radius=1.5, explode=[0, 0, 0, 0.1, 0.2],
       startangle=45, counterclock=True)
plt.show();
```
