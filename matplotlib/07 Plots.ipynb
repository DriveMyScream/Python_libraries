```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.arange(0, 10)
b = np.arange(11, 21)
```

```python
plt.scatter(a, b, label=['X label', 'Y label'])
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("The title")
plt.legend();
```

```python
plt.subplot(2, 2, 1)
plt.plot(a, b, 'r--')
plt.subplot(2, 2, 2)
plt.plot(a, b, 'g*--')
plt.subplot(2, 2, 3)
plt.plot(a, b, 'bo')
plt.subplot(2, 2, 4)
plt.plot(a, b, 'go');
```
