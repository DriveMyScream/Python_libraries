### treating the outlier using persentile

```python
lower_limit, upper_limit = df["price_per_sqft"].quantile([0.001, 0.999])
```

```python
newdf = df[(df["price_per_sqft"]>lower_limit) & (df["price_per_sqft"] < upper_limit)]
newdf
```

```python
newdf.shape
```

```python
outlier1 = df[(df["price_per_sqft"] < lower_limit) | (df["price_per_sqft"] > upper_limit)]
outlier1
```

#### so we are removing the total 196 number of outlier from the data set 

```python
plt.hist(newdf["price_per_sqft"], bins=20, rwidth=0.8)
plt.xlabel("price per sqft")
plt.ylabel("count")
plt.show()
```

### Outlier removing using 

```python
max_limit = newdf["price_per_sqft"].mean() + 4 * newdf["price_per_sqft"].std()
min_limit = newdf["price_per_sqft"].mean() - 4 * newdf["price_per_sqft"].std()
max_limit, min_limit
```

```python
newdf1 = newdf[(newdf["price_per_sqft"]>min_limit) & (newdf["price_per_sqft"] < max_limit)]
newdf1
```

```python
outlier2 = newdf[(newdf["price_per_sqft"] < min_limit) | (newdf["price_per_sqft"] > max_limit)]
outlier2
```

#### so we are removing the total 125 number of outlier from the data set

```python
plt.hist(newdf1["price_per_sqft"], bins=20, rwidth=0.8)
plt.xlabel("price per sqft")
plt.ylabel("Count")
plt.show()
```

### Remove with the help of zscore

```python
a = newdf1["price_per_sqft"]
b = newdf1["price_per_sqft"].mean()
c = newdf1["price_per_sqft"].std()
```

```python
newdf1["Z score"] = (a - b) / c
newdf1
```

```python
newdf2 = newdf1[(newdf1["Z score"]) > -4 & (newdf1["Z score"] < 4)]
newdf2
```

```python
outlier3 = newdf1[(newdf1["Z score"] < -4) | (newdf1["Z score"] > 4)]
outlier3
```

#### so we are removing the total 86 number of outlier from the data set

```python
plt.hist(newdf2["price_per_sqft"], bins=20, rwidth=0.8)
plt.xlabel("price per sqft")
plt.ylabel("count")
plt.show();
```
