```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Income data set")
```

```python
df
```

```python
df.describe()
```

```python
plt.hist(df["income($)"], bins=20, rwidth=0.8)
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()
```

```python
df["Log income"] = np.log(df["income($)"])
```

```python
df
```

```python
plt.hist(df["Log income"], bins=20, rwidth=0.8)
plt.xlabel("Log income")
plt.ylabel("Count")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
g = sns.barplot(x='income($)', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
g = sns.barplot(x='income($)', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
g.set(xscale="log");
```
