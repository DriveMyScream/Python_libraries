```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset("iris")
```

```python
df.sample(10)
```

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

```python
sns.jointplot(x=df['sepal_length'], y=df['sepal_width'], kind='reg');
```

```python
sns.jointplot(x=df['petal_length'], y=df['petal_width'], kind='reg');
```

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue='species');
```

```python
plt.subplot(2, 2, 1)
sns.distplot(df['sepal_length'], kde=True)
plt.subplot(2, 2, 2)
sns.distplot(df['sepal_width'], kde=True)
plt.subplot(2, 2, 3)
sns.distplot(df['petal_length'], kde=True)
plt.subplot(2, 2, 4)
sns.distplot(df['petal_width'], kde=True);
```

```python
sns.countplot(df['species']);
```

```python
plt.subplot(2, 2, 1)
sns.barplot(df["sepal_length"]);
plt.subplot(2, 2, 2)
sns.barplot(df['sepal_width']);
plt.subplot(2, 2, 3)
sns.barplot(df['petal_length']);
plt.subplot(2, 2, 4)
sns.barplot(df['petal_width']);
```

```python
plt.subplot(2, 2, 1)
sns.boxplot(df['sepal_length'])
plt.subplot(2, 2, 2)
sns.boxplot(df['sepal_width'])
plt.subplot(2, 2, 3)
sns.boxplot(df['petal_length'])
plt.subplot(2, 2, 4)
sns.boxplot(df['petal_width']);
```
