```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import matplotlib
```

```python
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
sns.set_style("darkgrid")
```

```python
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df.sample(10)
```

```python
flowers_df['species'].unique()
```

```python
plt.plot(flowers_df['sepal_length'], flowers_df['sepal_width']);
```

```python
sns.scatterplot(x=flowers_df['sepal_length'], y=flowers_df['sepal_width']);
```

```python
sns.scatterplot(x=flowers_df['sepal_length'], y=flowers_df['sepal_width'], hue=flowers_df['species'], s=100);
```

```python
sns.scatterplot(x=flowers_df['petal_length'], y=flowers_df['petal_width'], hue=flowers_df['species'], s=100);
```

```python
plt.figure(figsize=(12, 6))
plt.xlabel("The Sepal length")
plt.ylabel("The Petal length")
plt.title("The sepal length and petal length")
sns.scatterplot(x='sepal_length', y='petal_length', data=flowers_df, s=100, hue='species')
plt.grid();
```
