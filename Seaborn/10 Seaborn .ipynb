```python
def sineplot():
    x = np.linspace(0, 14, 100)
    for i in range(2):
        plt.plot(x, np.sin(x))
```

```python
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x+i*0.5) * (7-i))
```

```python
sineplot()
```
