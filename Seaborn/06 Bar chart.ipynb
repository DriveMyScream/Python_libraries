```python
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import numpy as np
```

```python
import matplotlib
```

```python
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
sns.set_style('darkgrid');
```

```python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
plt.title("The Prooduction of apples")
plt.xlabel("Years")
plt.ylabel("Yield of apples")
plt.bar(years, apples)
plt.legend(['Apples']);
```

```python
plt.title("The production of fruits")
plt.xlabel("Years")
plt.ylabel("yield")
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples)
plt.legend(['Apples', 'Oranges']);
```

## Bar plot with averages

```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df
```

```python
plt.title("The total bill in different days")
sns.barplot(x='day', y='total_bill', data=tips_df);
```

```python
plt.title("The total bill in different days")
sns.barplot(x='day', y='total_bill', data=tips_df, hue='sex');
```

```python
plt.title("The Total bill in different days")
sns.barplot(y='day', x='total_bill', hue='sex', data=tips_df);
```
