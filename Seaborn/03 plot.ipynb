```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apple = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
plt.plot(years, yield_apple);
```

```python
plt.title("Yield of apple year by year")
plt.xlabel("Years")
plt.ylabel("Yield of apples")
plt.plot(years, yield_apple);
```

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896]
```

```python
plt.title("Yield of apple and orange")
plt.xlabel("Years")
plt.ylabel("Yield (tones per hectare)")
plt.plot(years, apples)
plt.plot(years, oranges);
```

```python
plt.title("Production of apple and orange")
plt.xlabel("Years")
plt.ylabel("Yield (tones per hectare)")
plt.plot(years, apples, label="Apple")
plt.plot(years, oranges, label="Oranges")
plt.legend();
```

```python
plt.title("Production of apples and orange")
plt.xlabel("Years")
plt.ylabel("Yield of apples and oraange")
plt.plot(years, apples, label='Apple', marker="o")
plt.plot(years, oranges, label='Orange', marker="x")
plt.legend();
plt.grid();
```

```python
plt.title("The yeild of apples and oranges")
plt.xlabel("Years")
plt.ylabel("Yeild")
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=0.5)
plt.legend(['Apple', 'Orange'])
plt.grid();
```

#### fmt = '[marker][line][color]'

```python
plt.title("The yield of apples and oranges")
plt.xlabel("Days")
plt.ylabel("Yield")
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')
plt.legend(['Apples', 'Oranges'])
plt.grid();
```

```python
plt.title("yield oranges (tones per hectare)")
plt.xlabel("Years")
plt.ylabel("Yield")
plt.plot(years, oranges, 'or');
plt.legend(['Orange']);
```
