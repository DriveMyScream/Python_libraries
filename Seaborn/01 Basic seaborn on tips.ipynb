```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head(10)
```

#### Correlation with heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

## Join plot

### Univeriate analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex');
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg');
```

### Pair plot

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue='sex');
```

### Dist plot: it is used to cheack the distrubution of column features

```python
sns.distplot(df['total_bill']);
```

```python
sns.distplot(df['tip'], kde=True, bins=10);
```

## Categorical plots

### 1) Count plot

```python
sns.countplot(df['sex']);
```

```python
sns.countplot(y=df['sex']);
```

### 2) Bar plot: compare things between different groups or track changes over the time

```python
sns.barplot(x=df['total_bill'], y=df['sex']);
```

```python
sns.barplot(df['sex'], df['total_bill']);
```

### 3) Box plot

```python
sns.boxplot(df['smoker'], df['total_bill']);
```

```python
sns.boxplot(df['day'], df['total_bill']);
```

```python
sns.boxplot(data=df, orient='v');
```

```python
sns.boxplot(x=df['total_bill'], y=df['day'], hue=df['smoker']);
```

### 3) Violin plot

```python
sns.violinplot(df['total_bill'], df['day']);
```
